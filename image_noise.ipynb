```python
from numpy import random as r
from matplotlib import pyplot as p
```

```python
im = r.rand(256,256)
p.imshow(im, cmap='gray') #グレイ画像表示する
```

```python
im[0:64, 0:32]=0 #縦（y)が0~64, 横（x)が0~32の画素をすべて0にする
p.imshow(im, cmap='gray')
```

```python
im = r.rand(256,256,3)
p.imshow(im) #デフォルトはカラー画像を表示する
```
